**Since Edtech Dataset was not avaialble I used hospital dataset**

```python
import pandas as pd
import numpy as np
df=pd.read_csv('healthcare_dataset.csv')
df
```

```python
df.columns
```

**Q.1 Most Common Disease**

```python
print(df['Medical Condition'].value_counts().idxmax())
```

**Q.2 Avg Treatement cost per disease**

```python
print(df.groupby('Medical Condition')['Billing Amount'].mean())
```

**Q.3 Patients per hospital branch**

```python
print(df['Hospital'].value_counts())
```

**Q.4 Patients Older than 60**

```python
print(df[df['Age']>60]['Name'])
```

**Q.5 Branch with highest treatment cost**

```python
print(df.sort_values('Billing Amount',ascending=False)['Hospital'].head(1)) #considering the highest bill of an individual
print(df.groupby('Hospital')['Billing Amount'].sum().idxmax())  #if we're considering the sum of all paitents bill and then finding the highest
```

**Q.6 Avg age of Patients per disease**

```python
print(df.groupby('Medical Condition')['Age'].mean())
```

**Q.7 Top 5 highest treatment cost patients**

```python
print(df.sort_values('Billing Amount')['Name'].head(5))
```

**Q.8 Patients By Gender**

```python
print(df.groupby('Gender')['Gender'].value_counts())
```

**Q.9 Disease with highest treatment cost**

```python
print(df.sort_values('Billing Amount',ascending=False)['Medical Condition'].head(1)) #if we're  finding single treatment cost which is the highest
print(df.groupby('Medical Condition')['Billing Amount'].sum().idxmax())  #if we're considering the total treatment cost of a disease
```

**Q.10 Monthly Patients admissions**

```python
df['Date of Admission']=pd.to_datetime(df['Date of Admission'])
df.groupby(df['Date of Admission'].dt.month).size()
```
